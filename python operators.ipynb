```python
# 4.] Membership Operator
# Q7. Given a list of subjects:
# "Python Java C C++ Operating System UI/UX"
# Accept a subject name from the user
# Check whether it exists in the list using a membership operator
# Print appropriate message

subject = ['Python','java','c','c++','operating system','UI/UX']

subjects = input(" Enter a subject name:")

if subject.lower() in [s.lower() for s in subjects:
    print(f"{subject} exists in the list.")
else:
    print(f"{subject} does not exist in the list")
```
