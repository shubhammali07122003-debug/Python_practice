```python
# 5.] Accept any 5 digit number from the user, reverse it 
# and check whether the original and reverse are 
# same or not. check it palindrome or not 

num = int(input("Enter your number"))

original = num

n = num % 10
num = num // 10
n2 = num % 10
num = num // 10
n3 = num % 10
num = num // 10
n4 = num % 10
num = num // 10
n5 = num % 10

rev = n*10000 + n2*1000 + n3*100 + n4*10 + n5
print("Revrse:" , rev)
print("Same or not", original == rev )

```

```python
# 6.] Accept any year from the user and check whether 
# it is a leap year or not.(A leap year is a year which is 
# divisible by 4 but not by 100 and should be 
# divisible by 400).
year = int(input("Enter year:"))

is_leap = (year % 4 == 0 and year % 100 != 0) or ( year % 400 == 0)
print("Leap Year :",  is_leap)
```

```python
# 8.] Accept inner and outer radius of a ring and print 
# the perimeter and area of the ring 
# (Hint: perimeter = 2 π (a+b) , area = π (a*a-b*b) )
import math

a = float(input("Enter inner radius:"))
b = float(input("Enter outer radius:"))

perimeter = 2 * math.pi *(a+b)
area = math.pi * (a*a - b*b)

print("Perimeter:", perimeter)
print("Area:",area)
```

```python
# 10.] Write a program to convert specified days into years, 
# weeks and days. Note: Ignore leap year. assumes that all 
# months have 30 days and all years have 365 days
# i/p : Number of days : 1329
# Expected Output :  Years: 3    Weeks: 33      Days: 3

days = int(input("Enter days:"))

years = days // 365
days = days % 365

weeks = days // 7
days = days % 7

print("Years:",years)
print("Weeks:",weeks)
print("Days:",days)
```
